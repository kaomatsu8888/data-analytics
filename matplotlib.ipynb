```python
# 東京、那覇、札幌の気温
%matplotlib inline
import pandas as pd #pandasをインポート
import matplotlib.pyplot as plt #matplotlibをインポート
import seaborn as sns #seabornをインポート
sns.set(font=["Meiryo", "Yu Gothic", "Hiragino Maru Gothic Pro"]) #フォントを設定

data = {
    "月" : ["1月","2月","3月","4月","5月","6月","7月","8月","9月","10月","11月","12月"], #月
    "東京" : [5.6,7.2,10.6,13.6,20.0,21.7,25.8,27.3,23.3,18.3,13.9,8.8], #東京
    "那覇" : [17.5,17.5,19.6,21.9,24.7,26.9,28.9,29.5,28.0,25.7,22.5,19.4],
    "札幌" : [-3.6,-2.9,1.2,7.3,12.8,17.0,21.5,22.7,18.6,12.0,5.5,-0.1]
    }
df = pd.DataFrame(data) #データフレームを作成
df.head() #データフレームの先頭5行を表示
```

```python
df.plot()
plt.title("日本の気温の変化")
plt.show()
```

```python
# クッキー、ケーキが好きか嫌いかアンケート
%matplotlib inline
import pandas as pd #pandasをインポート
import matplotlib.pyplot as plt #matplotlibをインポート
import seaborn as sns #seabornをインポート

data = {
    "クッキー" : [35,47,18],
    "ケーキ" : [62,26,12]
    }
idx = ["好き","普通","嫌い"]
df = pd.DataFrame(data, index=idx) #データフレームを作成
df
```

```python
df["クッキー"].plot.pie()
plt.title("お菓子の好き嫌いは、どのような割合か？")
plt.show()
```

```python
df["クッキー"].plot.pie(startangle=90, counterclock=False, labeldistance=0.5)
plt.title("お菓子の好き嫌いは、どのような割合か？")
plt.show()
```

```python
df["ケーキ"].plot.pie(startangle=90, counterclock=False, labeldistance=0.5)
plt.title("お菓子の好き嫌いは、どのような割合か？")
plt.show()
```

```python
#身長、クラスデータ比較
%matplotlib inline
import pandas as pd #pandasをインポート
import matplotlib.pyplot as plt #matplotlibをインポート
import seaborn as sns #seabornをインポート
sns.set(font=["Meiryo", "Yu Gothic", "Hiragino Maru Gothic Pro"]) #フォントを設定

data = {
    "Aクラス" : [163.6,172.6,163.7,167.1,169.9,173.9,170.1,166.2,176.7,165.4],
    "Bクラス" : [166.9,172.7,166.4,173.4,169.6,171.8,166.9,168.2,166.7,169.8]
    }
df = pd.DataFrame(data) #データフレームを作成
df.head() #データフレームの先頭5行を表示
```

```python
sns.boxplot(data=df, width=0.2)
plt.title("身長のばらつきに違いはあるか？")
plt.show()
```
