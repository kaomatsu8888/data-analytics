```python
# ゴルトンボード
%matplotlib inline
import random
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font=["Meiryo", "Yu Gothic", "Hiragino Maru Gothic Pro"])

# ゴルトンボード表示関数：段数、球数を指定する
def goltun(steps, count) :
    # 玉が落ちた位置をいれる空のリストを作成
    ans = []
    # 指定された球数だけ繰り返す
    for i in range(count) :
        # 玉を落とす最初の位置を50にする
        val = 50
        # 指定された段数だけ繰り返す
        for j in range(steps) :
            # 0か1をランダムで、0なら-1, 1なら+1する
            if random.randint(0,1) == 0 :
                val -= 1
            else :
                val += 1
        # 玉が落ちた位置をリストに追加
        ans.append(val)

    # データフレームを作成
    df = pd.DataFrame(ans)
    # 0列目のデータをヒストグラムで表示
    df[0].plot.hist()
    # タイトルを設定
    plt.title(str(steps)+"段" +str(count)+"個")
    plt.ylabel("")
    plt.show()

# ゴルトンボードを表示
goltun(1, 10000)
            
```

```python
%matplotlib inline
import pandas as pd #pandasをインポート
import matplotlib.pyplot as plt #matplotlibをインポート
import seaborn as sns #seabornをインポート
import numpy as np #numpyをインポート
from scipy.stats import norm #scipy.statsからnormをインポート
sns.set(font=["Meiryo", "Yu Gothic", "Hiragino Maru Gothic Pro"]) #フォントを設定

df = pd.DataFrame({
    "A" : np.random.randint(0,100,1500),
    "B" : np.random.normal(50,10,1500)
    })

sns.distplot(df["A"], fit=norm, fit_kws={"color":"red"}) #Aのヒストグラムを表示
plt.title("かたよりのないrandomな値")
plt.show()

sns.distplot(df["B"], fit=norm, fit_kws={"color":"red"}) #Bのヒストグラムを表示
plt.title("かたよりのあるrandomな値")
plt.show()
```

```python
from scipy.stats import norm

scorelist = [60, 70, 80] # 点数のリスト
for score in scorelist : # 点数のリストを繰り返す
    cdf = norm.cdf(x=score, loc=50, scale=10) # 累積分布関数
    print("偏差値", score, "は、上から", (1-cdf)*100, "%") # 結果を表示
```

```python
perlist = [0.1586, 0.02275, 0.00134]
for per in perlist :
    ppf = norm.ppf(q=(1-per), loc=50, scale=10) # 累積分布関数
    print("上から", per*100, "%の偏差値は", ppf, "以上が必要です") # 結果を表示
```

```python
#IQの平均と標準偏差
from scipy.stats import norm

std = 15 # 標準偏差
IQlist = [110, 130, 148] # IQのリスト
for IQ in IQlist : # IQのリストを繰り返す
    cdf = norm.cdf(IQ, loc=100, scale=std) # 累積分布関数
    print("IQ", IQ, "は、上から", (1-cdf)*100, "%") # 結果を表示
```
