```python
# 数学、理科、社会の相関係数を求める
%matplotlib inline
import pandas as pd # データ分析ライブラリ
import matplotlib.pyplot as plt # グラフ描画ライブラリ
import seaborn as sns # グラフ描画ライブラリ
sns.set(font=["Yu Gothic", "Hiragino Maru Gothic Pro"]) # 日本語フォント設定

data = {
    "数学": [100, 85, 90, 95, 80, 80, 75, 65, 65, 60, 55, 45, 45],
    "理科": [94, 90, 95, 90, 85, 80, 75, 70, 60, 60, 50, 50, 48],
    "社会": [80, 88, 70, 62, 86, 70, 79, 65, 75, 67, 75, 68, 60]
}
df = pd.DataFrame(data)
df.head()
```

```python
df.plot.scatter(x="数学", y="理科", c="b")
plt.title("数学と理科の関係性")
plt.show()

df.plot.scatter(x="数学", y="社会", c="b")
plt.title("数学と社会の関係性")
plt.show()
```

```python
# 相関係数を求める(相関係数は-1から1までの値をとり、1に近いほど正の相関が強く、-1に近いほど負の相関が強い)
print("数学と理科 =", df.corr()["数学"]["理科"])
print("数学と社会 =", df.corr()["数学"]["社会"])
```

```python
print(df.corr()) # 相関係数をまとめて求める
```

```python
sns.regplot(data=df, x="数学", y="理科", line_kws={"color": "red"}) # 回帰直線を描画
plt.show() # 回帰直線を描画

sns.regplot(data=df, x="数学", y="社会", line_kws={"color": "red"}) # 回帰直線を描画
plt.show() # 回帰直線を描画
```

```python
# ヒストグラム月散布図 ＋ 回帰直線
sns.jointplot(data=df, x="数学", y="理科", kind="reg", line_kws={"color": "red"}) # 回帰直線を描画
plt.show() # 回帰直線を描画

sns.jointplot(data=df, x="数学", y="社会", kind="reg", line_kws={"color": "red"}) # 回帰直線を描画
plt.show() # 回帰直線を描画
```

```python
print(df.corr())
```

```python
# ヒートマップ
sns.heatmap(df.corr(), annot=True, vmax=1, vmin=-1, center=0) # annot=Trueで数値を表示
plt.show()
```
